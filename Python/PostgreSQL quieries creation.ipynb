```python
check_cwd = False
check_temptable = False

for dirpath, dirname, filenames in os.walk(os.getcwd()):

    if check_cwd == False:
        with open(os.getcwd() + '\\' + filenames[1], 'r') as f:
            index = f.readline().split(',')
        check_cwd = True


    for file in filenames:
        if check_temptable == False:
            print('CREATE TEMP TABLE coltable (')
            for i, col in enumerate(index):
                if i<3:
                    print('"' + col.strip() + '" varchar(100),')
                elif i<len(index) - 1:
                    print('"' + col.strip() + '" numeric,')
                else:
                    print('"' + col.strip() + '" numeric')
            print(');')
            check_temptable = True

        print("\nCREATE TABLE IF NOT EXISTS", '"' + file[:-4].lower().replace(' ', '_').replace(',', '') + '"', 'as SELECT * FROM coltable;')
        

        print('\n\nCOPY', '"' + file[:-4].lower().replace(' ', '_').replace(',', '') + '"',
                "\nFROM ", ("'" + os.getcwd() + '\\' + file + "'").replace('\\', '/'),
                "\nDELIMITER ','\nCSV HEADER;\n")
    
```
